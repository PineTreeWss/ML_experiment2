```python
# %load SVM2_ALL.py
#本代码用于完成机器学习实验一中SVM线性分类的操作
#by 王松盛
from sklearn.datasets import load_svmlight_file
from sklearn.model_selection import train_test_split
from matplotlib.ticker import LogLocator,FormatStrFormatter
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg
import random
import math
import copy
#先读取数据集
def GetData():
    train_x,train_y = load_svmlight_file('./dataset/a9a.txt', 123)
    train_x = train_x.toarray()
    test_x,test_y = load_svmlight_file('./dataset/a9a_test.txt',123)
    test_x = test_x.toarray()
    return train_x,train_y,test_x,test_y
#初始化模型:随机初始化
def getInitData(seedNum):
    np.random.seed(seedNum)
    w = np.random.rand(124)
    return w
#输出线性分类的结果
def linearClassifier(x,w,b):
    #print ('w',w,'x',x,'b',b)
    #input()
    val = (w*x).sum()+b
    if val > 0:
        return 1,val
    else:
        return -1,val
#本代码用于完成机器学习实验一中SVM线性分类的操作
#by 王松盛
from sklearn.datasets import load_svmlight_file
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg
import random
import math
import copy
#先读取数据集
def GetData():
    train_x,train_y = load_svmlight_file('./dataset/a9a.txt', 123)
    train_x = train_x.toarray()
    test_x,test_y = load_svmlight_file('./dataset/a9a_test.txt',123)
    test_x = test_x.toarray()
    return train_x,train_y,test_x,test_y
#初始化模型:随机初始化
def getInitData(seedNum):
    np.random.seed(seedNum)
    w = np.random.rand(124)
    return w
#输出线性分类的结果
def linearClassifier(x,w,b):
    #print ('w',w,'x',x,'b',b)
    #input()
    val = (w*x).sum()+b
    if val > 0:
        return 1,val
    else:
        return -1,val
#输出w和b梯度
def GetGradient(x,y,w,c,start,batch):
    b = w[-1]
    w = np.array(w.tolist()[:-1])
    wGrad = np.array([0.0] * len(w))
    bGrad = 0.0
    for i in range(start,start + batch):
        temp = 1-y[i] *((w * x[i]).sum() + b)
        if temp >= 0:
            wGrad += w - c * y[i] * x[i]
            bGrad -= c * y.sum()
        else:
            wGrad += w
    wGrad /= batch
    bGrad /= batch
    return np.array(wGrad.tolist() + [b])
#梯度下降算法
def GradientDescent(x,y,w,c,steplen,n_iteration,batch):
    w_Group = [w]
    for i in range(n_iteration):
        #if i % 100 == 0:
           # print (i)
        start = int(random.random() * (len(x) - batch))
        w = np.copy(w_Group[-1])
        wGrad = GetGradient(x,y,w,c,start,batch)
        w -= wGrad * steplen
        # temp = wGrad.tolist() + [bGrad]
        # temp = np.array(temp)
        # temp /= linalg.norm(temp) * steplen
        # w -= temp[:-1]
        # b -= temp[-1]
        w_Group.append(w)
    return w_Group
def GradientDescent_NAG(x,y,w,c,learningRate,n_iteration,batch,gamma):
    w_Group = [w]
    v = np.array([0.0]*124)
    for i in range(n_iteration):
        #if i % 1000 == 0:
            #print (i)
        index = int(random.random() * (len(x)-batch))
        w = np.copy(w)
        xGroup = []
        yGroup = []
        wGrad = GetGradient(x,y,w-gamma * v,c,index,batch)
        v = gamma * v + learningRate * wGrad
        # wGrad1,bGrad1 = GetGradient1(x,y,w,b)
        # if bGrad != bGrad1:
        #     print (i)
        #     print (bGrad,bGrad1)
        #     print (x,y,w,b)
        #     input()
        w -= v
        w_Group.append(w)
    return w_Group
def GradientDescent_RMSProp(x,y,w,c,learningRate,n_iteration,batch,gamma,ita):
    w_Group = [w]
    Gt = 0
    for i in range(n_iteration):
        #if i % 1000 == 0:
            #print (i)
        index = int(random.random() * (len(x)-batch))
        w = np.copy(w)
        xGroup = []
        yGroup = []
        wGrad = GetGradient(x,y,w,c,index,batch)
        Gt = gamma * Gt + (1-gamma)* wGrad.dot(wGrad)
        w -= learningRate / math.sqrt(Gt + ita) * wGrad
        w_Group.append(w)
    return w_Group
def GradientDescent_AdaDelta(x,y,w,c,n_iteration,batch,gamma,ita):
    w_Group = [w]
    Gt = 0.0
    delta = 0.0
    for i in range(n_iteration):
        #if i % 1000 == 0:
            #print (i)
        index = int(random.random() * (len(x)-batch))
        w = np.copy(w)
        wGrad = GetGradient(x,y,w,c,index,batch)
        Gt = gamma * Gt + (1-gamma)* wGrad.dot(wGrad)
        deltaTheta = - (math.sqrt(delta+ita) / math.sqrt(Gt + ita)) * wGrad
        w += deltaTheta
        delta = gamma * delta + (1-gamma) * deltaTheta.dot(deltaTheta)
        w_Group.append(w)
    return w_Group
def GradientDescent_Adam(x,y,w,c,learningRate,n_iteration,batch,gamma,beta,ita):
    w_Group = [w]
    Gt = 0
    m = np.array([0.0]*len(w))
    for i in range(n_iteration):
        #if i % 1000 == 0:
            #print (i)
        index = int(random.random() * (len(x)-batch))
        w = np.copy(w)
        wGrad = GetGradient(x,y,w,c,index,batch)
        m = beta * m + (1-beta) * wGrad
        Gt = gamma * Gt + (1-gamma)* wGrad.dot(wGrad)
        a = learningRate * (math.sqrt(1.0 - pow(gamma,i))/ (1.0 - pow(beta,i+1)))
        w -= a * m / math.sqrt(Gt + ita)
        w_Group.append(w)
    return w_Group
def loss(x,y,w,c):
    sumWr = 0
    b = w[-1]
    w = np.array(w.tolist()[:-1])
    for i in range(len(x)):
        temp = 1 - y[i] * ((w * x[i]).sum() + b)
        if temp <= 0:
            continue
        else:
            sumWr += temp
    return ((w*w).sum()*0.5 + c* sumWr) / len(x)
def Precision(x,y,w):
    b = w[-1]
    w = np.array(w.tolist()[:-1])
    right = 0
    for i in range(len(x)):
        ans = linearClassifier(x[i],w,b)[0]
        if ans == y[i]:
            right += 1
    return float(right) / len(x)
#主程序
def main(c,iteration,seed,batch):
    random.seed(seed)
    train_x,train_y,test_x,test_y = GetData()
    w = getInitData(seed)
    w_group = GradientDescent(train_x,train_y,w,c,0.002,iteration,batch)
    w_group_NAG = GradientDescent_NAG(train_x,train_y,w,c,0.002,iteration,batch,0.9)
    w_group_RMSProp = GradientDescent_RMSProp(train_x, train_y, w, c, 0.002, iteration, batch, 0.9, 0.0000001)
    w_group_AdaDelta = GradientDescent_AdaDelta(train_x, train_y, w, c, iteration, batch, 0.95, 0.00001)
    w_group_Adam = GradientDescent_Adam(train_x, train_y, w, c, 0.002, iteration, batch, 0.999, 0.9, 0.001)
    test_dif_Group = []
    test_dif_Group_NAG = []
    test_dif_Group_RMSProp = []
    test_dif_Group_AdaDelta = []
    test_dif_Group_Adam = []
    for i in range(len(w_group)):
        #print ('w',w_group[i])
        #print ('b',b_Group[i])
        if i % 10 != 0:
            continue
        test_dif_Group.append(loss(test_x,test_y,w_group[i],c))
        test_dif_Group_NAG.append(loss(test_x,test_y,w_group_NAG[i],c))
        test_dif_Group_RMSProp.append(loss(test_x, test_y, w_group_RMSProp[i], c))
        test_dif_Group_AdaDelta.append(loss(test_x, test_y, w_group_AdaDelta[i], c))
        test_dif_Group_Adam.append(loss(test_x, test_y, w_group_Adam[i], c))
        #print (train_dif_Group[-1])
        #print (test_dif_Group[-1])
    plt.plot([i*10 for i in range(len(test_dif_Group_AdaDelta))], test_dif_Group_AdaDelta, label='AdaDelta')
    plt.plot([i*10 for i in range(len(test_dif_Group_NAG))],test_dif_Group_NAG,label = 'NAG')
    plt.plot([i*10 for i in range(len(test_dif_Group_RMSProp))], test_dif_Group_RMSProp, label='RMSProp')
    plt.plot([i*10 for i in range(len(test_dif_Group))], test_dif_Group, label='SGD')
    plt.plot([i*10 for i in range(len(test_dif_Group_Adam))], test_dif_Group_Adam, label='Adam')
    #print (Precision(test_x,test_y,w_group[-1]))
    plt.yscale('log')
    plt.legend()
    plt.show()
main(0.958,4000,0,30)
```
