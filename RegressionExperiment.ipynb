```python
# %load logistic_regression.py
#本代码用于完成机器学习实验一中线性回归的操作
#by 王松盛
from sklearn.datasets import load_svmlight_file
from sklearn.model_selection import train_test_split
import numpy as np
from numpy import linalg
import matplotlib.pyplot as plt
import math
import random
#先获取数据集
def GetData():
    train_x,train_y = load_svmlight_file('./dataset/a9a.txt', 123)
    train_x = train_x.toarray()
    test_x,test_y = load_svmlight_file('./dataset/a9a_test.txt',123)
    test_x = test_x.toarray()
    return train_x,train_y,test_x,test_y
#初始化模型:0初始化
def getInitData():
    w = np.random.rand(124)
    return w
#输出线性分类的结果
def LogisticRegression(x,w):
    temp = -(w.dot(x))
    return 1.0/(1+math.e**temp)
#梯度下降算法
def GetGradient(x,y,w,i,batch,lamda):
    sum = np.array([0.0]*124)
    for i in range(i,i+batch):
        sum += y[i]*x[i]/(1+math.e**(y[i]*w.dot(x[i])))
    wGradient = sum / batch + lamda*w
    return wGradient
def GradientDescent(x,y,w,lamda,learningRate,n_iteration,seed,batch):
    w_Group = [w]
    random.seed(seed)
    for i in range(n_iteration):
        #if i % 1000 == 0:
            #print (i)
        index = int(random.random() * (len(x)-batch))
        w = np.copy(w)
        xGroup = []
        yGroup = []
        wGrad = GetGradient(x,y,w,index,batch,lamda)
        # wGrad1,bGrad1 = GetGradient1(x,y,w,b)
        # if bGrad != bGrad1:
        #     print (i)
        #     print (bGrad,bGrad1)
        #     print (x,y,w,b)
        #     input()
        w -= wGrad *learningRate
        w_Group.append(w)
    return w_Group
def GradientDescent_NAG(x,y,w,lamda,learningRate,n_iteration,seed,batch,gamma):
    w_Group = [w]
    random.seed(seed)
    v = np.array([0.0]*124)
    for i in range(n_iteration):
        # if i % 1000 == 0:
           # print (i)
        index = int(random.random() * (len(x)-batch))
        w = np.copy(w)
        xGroup = []
        yGroup = []
        wGrad = GetGradient(x,y,w-gamma * v,index,batch,lamda)
        v = gamma * v + learningRate * wGrad
        # wGrad1,bGrad1 = GetGradient1(x,y,w,b)
        # if bGrad != bGrad1:
        #     print (i)
        #     print (bGrad,bGrad1)
        #     print (x,y,w,b)
        #     input()
        w -= v
        w_Group.append(w)
    return w_Group
def GradientDescent_RMSProp(x,y,w,lamda,learningRate,n_iteration,seed,batch,gamma,ita):
    w_Group = [w]
    random.seed(seed)
    Gt = 0
    for i in range(n_iteration):
        #if i % 1000 == 0:
            #print (i)
        index = int(random.random() * (len(x)-batch))
        w = np.copy(w)
        xGroup = []
        yGroup = []
        wGrad = GetGradient(x,y,w,index,batch,lamda)
        Gt = gamma * Gt + (1-gamma)* wGrad.dot(wGrad)
        w -= learningRate / math.sqrt(Gt + ita) * wGrad
        w_Group.append(w)
    return w_Group
def GradientDescent_AdaDelta(x,y,w,lamda,n_iteration,seed,batch,gamma,ita):
    w_Group = [w]
    random.seed(seed)
    v = np.array([0.0]*124)
    Gt = 0.0
    delta = 0.0
    for i in range(n_iteration):
       # if i % 1000 == 0:
            #print (i)
        index = int(random.random() * (len(x)-batch))
        w = np.copy(w)
        wGrad = GetGradient(x, y, w, index, batch, lamda)
        Gt = gamma * Gt + (1-gamma)* wGrad.dot(wGrad)
        deltaTheta = - (math.sqrt(delta+ita) / math.sqrt(Gt + ita)) * wGrad
        w += deltaTheta
        delta = gamma * delta + (1-gamma) * deltaTheta.dot(deltaTheta)
        w_Group.append(w)
    return w_Group
def GradientDescent_Adam(x,y,w,lamda,learningRate,n_iteration,seed,batch,gamma,beta,ita):
    w_Group = [w]
    random.seed(seed)
    Gt = 0
    m = np.array([0.0]*len(w))
    for i in range(n_iteration):
        #if i % 1000 == 0:
           # print (i)
        index = int(random.random() * (len(x)-batch))
        w = np.copy(w)
        wGrad = GetGradient(x,y,w,index,batch,lamda)
        m = beta * m + (1-beta) * wGrad
        Gt = gamma * Gt + (1-gamma)* wGrad.dot(wGrad)
        a = learningRate * (math.sqrt(1.0 - pow(gamma,i))/ (1.0 - pow(beta,i+1)))
        w -= a * m / math.sqrt(Gt + ita)
        w_Group.append(w)
    return w_Group
#计算数据集的Square loss
def loss(x,y,w,lamda):
    loss = 0.0
    for i in range(len(x)):
        temp = -y[i] * (w.dot(x[i]))
        loss += math.log(1+math.e**temp)
    loss /= len(x)
    loss += lamda / 2 * w .dot(w)
    return loss
#主程序
def main(iteration,seed,batch,lamda):
    train_x,train_y,test_x,test_y = GetData()
    w = getInitData()
    train_x = np.column_stack((train_x,np.array([1.0]*len(train_x)).T))
    test_x = np.column_stack((test_x, np.array([1.0] * len(test_x)).T))
    w_group = GradientDescent(train_x,train_y,w,lamda,0.01,iteration,seed,batch)
    w_group_NAG = GradientDescent_NAG(train_x,train_y,w,lamda,0.01,iteration,seed,batch,0.8)
    w_group_RMSProp = GradientDescent_RMSProp(train_x,train_y,w,lamda,0.01,iteration,seed,batch,0.9,0.00000001)
    w_group_AdaDelta = GradientDescent_AdaDelta(train_x, train_y, w, lamda, iteration, seed, batch, 0.95, 0.00001)
    w_group_Adam = GradientDescent_Adam(train_x,train_y,w,lamda,0.01,iteration,seed,batch,0.999,0.9,0.001)
    test_dif_Group = []
    test_dif_Group_NAG = []
    test_dif_Group_RMSProp = []
    test_dif_Group_AdaDelta = []
    test_dif_Group_Adam = []
    for i in range(len(w_group)):
        if i % 10 != 0:
            continue
        #print ('w',w_group[i])
        #print ('b',b_Group[i])
        test_dif_Group.append(loss(test_x,test_y,w_group[i],lamda))
        test_dif_Group_NAG.append(loss(test_x, test_y, w_group_NAG[i], lamda))
        test_dif_Group_RMSProp.append(loss(test_x,test_y,w_group_RMSProp[i],lamda))
        test_dif_Group_AdaDelta.append(loss(test_x,test_y,w_group_AdaDelta[i],lamda))
        test_dif_Group_Adam.append(loss(test_x, test_y, w_group_Adam[i], lamda))
        #print (train_dif_Group[-1])
        #print (test_dif_Group[-1])
    #print (train_dif_Group)
    #print (test_dif_Group)
    plt.plot([i*10 for i in range(len(test_dif_Group_NAG))],test_dif_Group_NAG, label = 'NAG_loss')
    plt.plot([i*10 for i in range(len(test_dif_Group))], test_dif_Group, label= 'SGD_loss')
    plt.plot([i*10 for i in range(len(test_dif_Group))], test_dif_Group_RMSProp, label='RMSProp_loss')
    plt.plot([i*10 for i in range(len(test_dif_Group))], test_dif_Group_AdaDelta, label='AdaDelta_loss')
    plt.plot([i*10 for i in range(len(test_dif_Group))], test_dif_Group_Adam, label='Adam_loss')
    plt.yscale('log')
    plt.legend()
    plt.show()
main(2000,200,30,1)
```
